```python
import pandas as pd
from sqlalchemy import create_engine 
```

```python
teams_file = "Teams.csv"
teams_df = pd.read_csv(teams_file)
teams_df.head()
```

```python
teams_df = teams_df.rename(columns={"Team_ID": "id",
                                    "Team_Name": "Team"})
teams_df.head()
```

```python
results_file = "results.csv"
results_df = pd.read_csv(results_file)
results_df.head()
```

```python
tourney_file = "tourney.csv"
tourney_df = pd.read_csv(tourney_file)
tourney_df.columns
```

```python
results_columns= results_df[['Wteam_name', 'Lteam_name', 'Season', 'Wscore', 'Lscore', 'Wfgm', 'Wfga', 'Wfgm3', 'Wfga3',
       'Wftm', 'Wfta', 'Wor', 'Wdr', 'Wast', 'Wto', 'Wstl', 'Lfgm', 'Lfga', 'Lfgm3', 'Lfga3', 'Lftm', 'Lfta', 'Lor', 'Ldr', 'Last',
       'Lto', 'Lstl']].copy()

results_clean = results_columns.rename(columns={"Wteam_name": "winning_team",
                                                "Lteam_name": "losing_team",
                                                "Season": "season",
                                               "Wscore": "winning_score",
                                               "Lscore": "losing_score"
                                               })

results_clean.head()
```

```python
results_clean.to_csv('results_clean.csv')
```

```python
tourney_columns= tourney_df[['Wteam_name', 'Lteam_name', 'Season', 'Wscore',
       'Lteam', 'Wfgm', 'Wfga', 'Wfgm3', 'Wfga3',
       'Wftm', 'Wfta', 'Wor', 'Wdr', 'Wast', 'Wto', 'Wstl', 'Lfgm', 'Lfga', 'Lfgm3', 'Lfga3', 'Lftm', 'Lfta', 'Lor', 'Ldr', 'Last',
       'Lto', 'Lstl']].copy()

tourney_clean = tourney_columns.rename(columns={"Wteam_name": "winning_team",
                                                "Lteam_name": "losing_team",
                                                "Season": "season",
                                               "Wscore": "winning_score",
                                               "Lscore": "losing_score"})

tourney_clean.head()
```

```python
tourney_clean.to_csv('tourney_clean.csv')
```

```python
results_new_clean = "results_new_clean.csv"
results_df = pd.read_csv(results_new_clean)
results_df.head()
```

```python
tourney_new_clean = "tourney_new_clean.csv"
tourney_df = pd.read_csv(tourney_new_clean)
tourney_df.head()
```

```python
connection_string = "postgres:Mrtubbs10@localhost:5432/march_madness2.0"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
teams_df.to_sql(name='teams', con=engine, if_exists='append', index=True)
```

```python
results_df.to_sql(name='results', con=engine, if_exists='append', index=True)
```

```python
tourney_df.to_sql(name='tourney', con=engine, if_exists='append', index=True)
```

```python
conn = engine.connect()

duke_2015_data = pd.read_sql("select * from results\
                    where season='2015' \
                    and winning_team='Duke'\
                    union\
                    select * from results\
                    where season='2015'\
                    and losing_team='Duke'\
                    union\
                    select * from tourney\
                    where season='2015'\
                    and winning_team='Duke'\
                    order by index asc", conn)
duke_2015_data
```

```python
wisconsin_2015_data = pd.read_sql("select * from results\
                    where season='2015' \
                    and winning_team='Wisconsin'\
                    union\
                    select * from results\
                    where season='2015'\
                    and losing_team='Wisconsin'\
                    union\
                    select * from tourney\
                    where season='2015'\
                    and winning_team='Wisconsin'\
                    union\
                    select * from tourney\
                    where season='2015'\
                    and losing_team='Wisconsin'\
                    order by index asc", conn)
wisconsin_2015_data
```

```python
duke_2015_data.to_csv("duke_2015_data.csv")
wisconsin_2015_data.to_csv("wisconsin_2015_data.csv")
```

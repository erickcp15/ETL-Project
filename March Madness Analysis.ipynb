```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
tourney_df = pd.read_csv("tourney_clean.csv")
tourney_df.head()
```

```python
tourney_df.info()
```

```python
 tourney_df['Winning_field_goal_percent'] = tourney_df['Wfgm']/tourney_df['Wfga']*100
 tourney_df['Winning_field_goal_percent'].head()
   
```

```python
tourney_df['Winning_3_percent'] = tourney_df['Wfgm3']/tourney_df['Wfga3']*100
tourney_df['Winning_3_percent'].head()                                         
```

```python
tourney_df['Winning_free_throw_percent'] = tourney_df['Wftm']/tourney_df['Wfta']*100
tourney_df['Winning_free_throw_percent'].head()
```

```python
tourney_df['Winning_total_rebounds'] = sum(tourney_df['Wor'],tourney_df['Wdr'])
tourney_df['Winning_total_rebounds'].head()
```

```python
tourney_df['Losing_field_goal_percent'] = tourney_df['Lfgm']/tourney_df['Lfga']*100
tourney_df['Losing_field_goal_percent'].head()
```

```python
tourney_df['Losing_3_percent'] = tourney_df['Lfgm3']/tourney_df['Lfga3']*100
tourney_df['Losing_3_percent'].head()   
```

```python
tourney_df['Losing_free_throw_percent'] = tourney_df['Lftm']/tourney_df['Lfta']*100
tourney_df['Losing_free_throw_percent'].head()   
```

```python
tourney_df['Losing_total_rebounds'] = sum(tourney_df['Lor'],tourney_df['Ldr'])
tourney_df['Losing_total_rebounds'].head()
```

```python
tourney_df2 = tourney_df[['season','winning_team','winning_score',\
         'losing_team','losing_score','Winning_field_goal_percent','Winning_3_percent','Winning_free_throw_percent'\
         ,'Winning_total_rebounds','Wast','Wto','Wstl','Losing_field_goal_percent','Losing_3_percent','Losing_free_throw_percent'\
        ,'Losing_total_rebounds','Last','Lto','Lstl']].copy()
```

```python
tourney_df2
```

```python
tourney_df2 = tourney_df2.rename(columns={'Wast':'Winning_assists','Wto':'Winning_turnovers','Wstl':'Winning_steals',\
                         'Last':'Losing_assists','Lto':'Losing_turnover','Lstl':'Losing_steals'})
tourney_df2.head()
```

```python
tourney_df2.to_csv('tourney_new_clean.csv')
```

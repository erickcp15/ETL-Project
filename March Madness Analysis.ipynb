```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

# 1. Cleaning Dataset : Tourney - Tournament Results

```python
# Read csv file and put into a dataframe
tourney_df = pd.read_csv("tourney.csv")
tourney_df.head()
```

```python
# Display info
tourney_df.info()
```

```python
# Read csv file and put into a dataframe
teams_df = pd.read_csv("teams.csv")
teams_df.head()
```

```python
teams_df.info()
```

```python
teams_df['Team_Name'].value_counts()
```

```python
tourney_df = tourney_df.rename(columns = {'Wteam':'Team_Id'})
tourney_df.head()
```

```python
# Creating new columns for the tournament dataframe based on calculations made using existing columns
# A. Winning Field Goal Percentage
tourney_df['Winning_field_goal_percent'] = (tourney_df['Wfgm']/tourney_df['Wfga']*100).round(2)
tourney_df['Winning_field_goal_percent'].head()
   
```

```python
# B. Winning 3 percentage 
tourney_df['Winning_3_percent'] = (tourney_df['Wfgm3']/tourney_df['Wfga3']*100).round(2)
tourney_df['Winning_3_percent'].head()                                         
```

```python
# C. Winning free throw percentage
tourney_df['Winning_free_throw_percent'] = (tourney_df['Wftm']/tourney_df['Wfta']*100).round(2)
tourney_df['Winning_free_throw_percent'].head()
```

```python
# D. Winning total rebounds
tourney_df['Winning_total_rebounds'] = (tourney_df['Wor']+tourney_df['Wdr'])
tourney_df['Winning_total_rebounds'].head()
```

```python
# E. Losing field goal percentage
tourney_df['Losing_field_goal_percent'] = (tourney_df['Lfgm']/tourney_df['Lfga']*100).round(2)
tourney_df['Losing_field_goal_percent'].head()
```

```python
# F. Losing 3 Percentage
tourney_df['Losing_3_percent'] = (tourney_df['Lfgm3']/tourney_df['Lfga3']*100).round(2)
tourney_df['Losing_3_percent'].head()   
```

```python
# G. Losing free throw percentage
tourney_df['Losing_free_throw_percent'] = (tourney_df['Lftm']/tourney_df['Lfta']*100).round(2)
tourney_df['Losing_free_throw_percent'].head()   
```

```python
# H. Losing Total Rebounds
tourney_df['Losing_total_rebounds'] = (tourney_df['Lor']+tourney_df['Ldr'])
tourney_df['Losing_total_rebounds'].head()
```

```python
# Merge on Team ID to bring over Winning Teams' Names
merged_tourney = pd.merge(tourney_df, teams_df, on = ['Team_Id'])
merged_tourney.head()
```

```python
# Create a new dataframe by incorporating the calculated columns and indentified columns of interest.
tourney_df2 = merged_tourney[['Season','Team_Name','Wscore',\
                              'Lteam','Lscore','Winning_field_goal_percent',\
                              'Winning_3_percent','Winning_free_throw_percent',\
                              'Winning_total_rebounds','Wast','Wto','Wstl',\
                              'Losing_field_goal_percent','Losing_3_percent',\
                              'Losing_free_throw_percent','Losing_total_rebounds',\
                              'Last','Lto','Lstl']].copy()
tourney_df2.head()
```

```python
# Rename columns of interest.
tourney_df2 = tourney_df2.rename(columns={'Season':'season','Team_Name':'winning_team','Wscore':'winning_score','Lteam':'losing_team','Lscore':'losing_score','Wast':'Winning_assists','Wto':'Winning_turnovers','Wstl':'Winning_steals',\
                         'Last':'Losing_assists','Lto':'Losing_turnover','Lstl':'Losing_steals'})
tourney_df2.head()
```

```python
tourney_df2.info()
```

```python
# Merge tourney_df2 and teams_df to bring over the losing team names using the Team ID column and losing team column.
merged_tourney2 = pd.merge(tourney_df2, teams_df, left_on = 'losing_team', right_on = 'Team_Id')
merged_tourney2.head()
```

```python
# Reposition columns and rename dataframe.
tourney_df3 = merged_tourney2[['season','winning_team','winning_score','Team_Name','losing_score','Winning_field_goal_percent','Winning_3_percent','Winning_free_throw_percent','Winning_total_rebounds','Winning_assists',\
                               'Winning_turnovers','Winning_steals','Losing_field_goal_percent','Losing_3_percent','Losing_free_throw_percent','Losing_total_rebounds','Losing_assists','Losing_turnover','Losing_steals']].copy()
tourney_df3.head()
```

```python
# Rename the column'Team_Name' to 'losing_team' and sort by season - ascending, winning score - descending, and winning team ascending. Reset the index and drop old index.
tourney_df3 = tourney_df3.rename(columns={'Team_Name':'losing_team'})
sorted = tourney_df3.sort_values(by = ['season','winning_score','winning_team'], ascending = [True,False,True])
#df.sort_values(['a', 'b'], ascending=[True, False])
tourney_df4 = sorted.reset_index(drop=True)
tourney_df4.head()
```

```python
# Save new dataframe to a csv file.
tourney_df4.to_csv('tourney_new_clean.csv')
```
